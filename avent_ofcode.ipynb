```python
lines = [
"Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green",
"Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue",
"Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red",
"Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red",
"Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green"
]
with open("input_day02.txt", 'r') as fp:
    lines = fp.readlines()
colors = ['blue', 'red', 'green']
colors_limits = {'blue':14, 'red':12, 'green':13}

valid_games = []
invalid_games = []
total_score = 0
for line in lines:
    game_is_valid = True
    game_splitted = line.split(':')
    #print(game_splitted)
    game_number = int(game_splitted[0].replace("Game ", ""))
    games_min_set = {k:0 for k in colors}
    print('_____________')
    print(game_number)
    game_grabs = game_splitted[1].strip().split(';')
    #print(game_grabs)
    for grab in game_grabs:
        play = {k:0 for k in colors}
        colors_and_numbers = grab.strip().split(',')
        for cn in colors_and_numbers:
            cn_split = cn.strip().split(' ')
            #print(cn_split)
            n = int(cn_split[0])
            c = cn_split[1]
            play[c] = n
            if n > colors_limits[c]:
                game_is_valid = game_is_valid and False
        print(play)
        for k, v in play.items():
            if v > games_min_set[k]:
                games_min_set[k] = v
    print('min set : ')
    print(games_min_set)
    game_score = 1
    for c in colors:
        game_score = game_score * games_min_set[c]
    print(game_score)
    total_score = total_score + game_score



print(total_score)    
```
