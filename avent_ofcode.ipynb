```python
lines = [
"Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green",
"Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue",
"Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red",
"Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red",
"Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green"
]
with open("input_day02.txt", 'r') as fp:
    lines = fp.readlines()
colors = ['blue', 'red', 'green']
colors_limits = {'blue':14, 'red':12, 'green':13}

valid_games = []
invalid_games = []

for line in lines:
    game_is_valid = True
    game_splitted = line.split(':')
    print(game_splitted)
    game_number = int(game_splitted[0].replace("Game ", ""))
    game_grabs = game_splitted[1].strip().split(';')
    print(game_grabs)
    for grab in game_grabs:
        play = {k:0 for k in colors}
        colors_and_numbers = grab.strip().split(',')
        for cn in colors_and_numbers:
            cn_split = cn.strip().split(' ')
            print(cn_split)
            n = int(cn_split[0])
            c = cn_split[1]
            play[c] = n
            if n > colors_limits[c]:
                game_is_valid = game_is_valid and False


    if game_is_valid:
        valid_games.append(game_number)
    else:
        invalid_games.append(game_number)

sum_valids = 0    
print(valid_games)
for i in valid_games:
    sum_valids = sum_valids + i
print(sum_valids)    
```

```python
lines = [
"Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green",
"Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue",
"Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red",
"Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red",
"Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green"
]
with open("input_day02.txt", 'r') as fp:
    lines = fp.readlines()
colors = ['blue', 'red', 'green']
colors_limits = {'blue':14, 'red':12, 'green':13}

valid_games = []
invalid_games = []
total_score = 0
for line in lines:
    game_is_valid = True
    game_splitted = line.split(':')
    #print(game_splitted)
    game_number = int(game_splitted[0].replace("Game ", ""))
    games_min_set = {k:0 for k in colors}
    print('_____________')
    print(game_number)
    game_grabs = game_splitted[1].strip().split(';')
    #print(game_grabs)
    for grab in game_grabs:
        play = {k:0 for k in colors}
        colors_and_numbers = grab.strip().split(',')
        for cn in colors_and_numbers:
            cn_split = cn.strip().split(' ')
            #print(cn_split)
            n = int(cn_split[0])
            c = cn_split[1]
            play[c] = n
            if n > colors_limits[c]:
                game_is_valid = game_is_valid and False
        print(play)
        for k, v in play.items():
            if v > games_min_set[k]:
                games_min_set[k] = v
    print('min set : ')
    print(games_min_set)
    game_score = 1
    for c in colors:
        game_score = game_score * games_min_set[c]
    print(game_score)
    total_score = total_score + game_score



print(total_score)    
```

```python
lines = [
"467.....1$",
"1..*......",
"..35..633.",
"......#...",
"617*......",
".....+.58.",
"..592.....",
"......755.",
"...$.*....",
".664.598..",
".......123",
"345*..+...",
"........23",
"345...+...",
"....3+....",
"..5...4...",
"...+......",
"....5....."]

expected = [467, 1, 35, 633, 617, 592, 755, 664, 598, 123, 345, 3, 5, 4, 5]

with open("input_day03.txt", 'r') as fp:
    lines = fp.readlines()

symbols_positions = []
numbers_positions = []
numbers_lines = []

for x, line in enumerate(lines):
    numbers_line = []
    current_number_positions = []
    current_number = ""
    catch_num = False
    for y, c in enumerate(line):
        # si le caractere n'est pas numerique 
        if not c.isnumeric():
            catch_num = False
        else:
            catch_num = True
        
        # pas numerique ni un point donc un symbole
        if not catch_num and c != '.':
            symbols_positions.append((x,y))
        # sinon un point ou un nombre
        else:
            if catch_num:
                current_number = "%s%s" % (current_number,c)
                current_number_positions.append((x,y))
        
        # si nous sommes sur le dernier caractere ou que celui en cours n'est pas numérique (symbole ou point)
        marqueur_fin_nombre = (not catch_num) or (y == len(line) - 1)
        if marqueur_fin_nombre:
            #si on vient tout juste de quitter un nombre
            if current_number != "":
                struc_number_position = {'number': int(current_number), 'position_x_y' : current_number_positions}
                numbers_line.append(struc_number_position)
                current_number = ""
                current_number_positions = []
    numbers_lines.append(numbers_line)
print(symbols_positions)
print(numbers_lines)


number_with_symbol_neighboor = []
number_without_symbol_neighboor = []
for line in numbers_lines:
    #pour chaque nombre de la ligne et son lot de position
    for struct_num_positions in line:
        has_found_symbol_in_neighboor = False
        for tuple_xy in struct_num_positions['position_x_y']:
            for tuple_symbol_xy in symbols_positions:
                # meme ligne 
                if (tuple_xy[0] == tuple_symbol_xy[0]) or (tuple_xy[0] == tuple_symbol_xy[0] + 1) or (tuple_xy[0] == tuple_symbol_xy[0] - 1):
                    # meme colone ou celle d'avant ou apres
                    if tuple_xy[1] == tuple_symbol_xy[1] or (tuple_xy[1] == tuple_symbol_xy[1]+1) or (tuple_xy[1] == tuple_symbol_xy[1]-1):
                        has_found_symbol_in_neighboor = True
        if has_found_symbol_in_neighboor:
            number_with_symbol_neighboor.append(struct_num_positions['number'])
        else:
            number_without_symbol_neighboor.append(struct_num_positions['number'])

                

all_numbers = [ struct_num_positions['number'] for line in numbers_lines for struct_num_positions in line]

print("len(lines) : %s" % len(lines))
print("all_numbers: %s " % all_numbers)
# with duplicates !
print("number_with_symbol_neighboor: %s" % number_with_symbol_neighboor)
print("expected: %s" % expected)

print("set(number_with_symbol_neighboor): %s" % set(number_with_symbol_neighboor))
print("set(expected): %s" % set(expected))
print("number_without_symbol_neighboor: %s" % number_without_symbol_neighboor)

sum = 0
#drop duplicates !
for k in set(number_with_symbol_neighboor):
    sum = sum + k
print("sum %s " % sum)
sum_wd = 0
#with duplicates !
for k in number_with_symbol_neighboor:
    sum_wd = sum_wd + k
print("sum %s " % sum_wd)

sum_expected = 0
for i in set(expected):
    sum_expected = sum_expected + i
print("sum expected %s " % sum_expected)

```
